```python
class Tienda_online:
    def __init__(self):
        self.inventario=[]

    def agregar_producto(self,nombre,precio,cantidad):
        for producto in self.inventario:
           if producto["nombre"] == nombre:
            producto["cantidad"] = producto["cantidad"] + cantidad
            return
        producto_nuevo = {
        "nombre": nombre,
        "precio": precio,
        "cantidad": cantidad
    }
    
        self.inventario.append(producto_nuevo)

    def informacion_inventario(self):
         for producto in self.inventario:
          print(producto["nombre"],
                producto["precio"],
                producto["cantidad"])
               
    
    
                 
                 
             
             
         
         


    
```

```python
tienda1=Tienda_online()
tienda1.agregar_producto("camisa",15,5)
tienda1.agregar_producto("top",20,3)
tienda1.agregar_producto("camisa",15,2)
tienda1.informacion_inventario()
```
